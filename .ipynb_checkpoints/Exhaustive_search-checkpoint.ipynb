```python
from matplotlib import pyplot as plt
import numpy as np 
from time import time
```

```python
U =lambda t: ((204165.5)/(330-2*t))+(10400/(t-20))
```

```python
t = np.linspace(40,90,500)
fu=U(t)
fig1=plt.figure(figsize = (8,5))
plt.plot(t,fu);
plt.xlabel('Temperature (t)')
plt.ylabel('Cost U(t)')
plt.grid(True)
```

```python
t = np.linspace(40,90,50000)
minU = 2**99
minT = 0


start_time = time()
for i in t:
    if U(i) < minU:
        minU=U(i)
        minT=i

elapsed_time = time() - start_time
elapsed_time = elapsed_time*1000
print("Tiempo transcurrido: %f ms." % elapsed_time )
print("x: %f" % minT)
print("fx: %f" % minU)
print("f'x: %f" % ((U(minT+0.01)-U(minT))/0.01))
print("iteraciones: %d" % len(t))
```
