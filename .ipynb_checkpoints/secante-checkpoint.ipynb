```python
import numpy as np
from matplotlib import pyplot as plt
from time import time
```

```python
f_a =lambda t:((204165.5)/(330-2*t))+(10400/(t-20))
dx = lambda x, delta: (f_a(x+delta)-f_a(x))/delta
```

```python
delta = 0.001
epsilon = delta
x1 = 90
x2 = 40
start_time = time()
for i in range(1,100):
    alpha = x2 - ((dx(x2,delta)*(x2-x1)))/(dx(x2,delta)-dx(x1,delta))
    if dx(alpha,delta)*dx(x1,delta)<0:
        x2=alpha
    else:
        x1 = alpha
    if abs(dx(alpha,delta))<epsilon:
        break

elapsed_time = time() - start_time
elapsed_time = elapsed_time*1000
print("Tiempo transcurrido: %f ms." % elapsed_time )
print("x: %f" % alpha)
print("fx: %f" % f_a(alpha))
print("f'x: %f" % dx(alpha,delta))
print("iteraciones: %d" % i)
```

```python
t = np.linspace(40,90,100)
fu=dx(t,delta)
fig, ax = plt.subplots(nrows=10, ncols=1,figsize=(10, 70)) 
x1 = 90
x2 = 40
for i in range(10):
    ax[i].plot(t,fu)
    ax[i].grid(True, which='both')
    ax[i].axhline(y=0, color='k')
    ax[i].plot(x1,dx(x1,delta), color="red", marker="o",  markersize=7)
    ax[i].plot(x2,dx(x2,delta), color="red", marker="o",  markersize=7)
    x_values = [x1,x2]
    y_values = [dx(x1,delta),dx(x2,delta)]
    ax[i].plot(x_values, y_values)
    alpha = x2 - ((dx(x2,delta)*(x2-x1)))/(dx(x2,delta)-dx(x1,delta))
    ax[i].plot(alpha,dx(alpha,delta), color="green", marker="o",  markersize=7)
    ax[i].set_title("x: %f fx: %f f'x: %f" % (alpha,f_a(alpha),dx(alpha,delta)))
    if dx(alpha,delta)*dx(x1,delta)<0:
        x2=alpha
    else:
        x1 = alpha
    if abs(dx(alpha,delta))<epsilon:
        break
```
