# <a id="biseccion">Bisección</a>


```python
import numpy as np
import math
import matplotlib.pyplot as plt

#funcion
f = lambda x: 204165.5/(330-2*x) + 10400/(x-20)
#derivada
fdx = lambda x,dx: (f(x+dx) - f(x-dx))/(2*dx)



def bisection(a,b,e):
    dx = 0.001
    for i in range (1,300):
        alpha = (a+b)/2
        d_alpha = fdx(alpha,dx)
        d_a = fdx(a,e)
        if d_alpha*d_a<0:
            b=alpha
        else:
            a=alpha
        if abs(a-b)<e:
            break
        print('a={0}, b={1},  U(T*)={2} ite={3} \n'.format(a,b,f(a),i))

    print('T*={0}  U(T*)= {1} iteraciones={2} \n'.format(a,f(a),i))
    return(a,i)


#temperatura
T = np.arange(40,90.001, 0.001)
e = math.exp(1)**-3

#funcion evaluada en T
fx = f(T)

Tdx =  fdx(T,e)


%time xmin,i =  bisection(40,90,e)
print("x*: " + str(xmin))
print("f(x*):"  + str(f(xmin)))
print("f(x'): " + str(fdx(xmin, 0.001)))
print("iteraciones: " + str(i))
```

# <a id="secante">Secante</a>


```python
import numpy as np
import math
import random

f = lambda x: 204165.5 / (330 - 2 * x) + 10400 / (x - 20)
fdx = lambda x, dx: (f(x + dx) - f(x - dx)) / (2 * dx)


def Seccant(x1, x2, fdx, epsilon, max_iter):
    dx = 0.001
    for i in range(0, max_iter):
        actualmin = x2 - ((fdx(x2, dx) * (x2 - x1))) / (fdx(x2, dx) - fdx(x1, dx))

        if abs(fdx(actualmin, dx)) < epsilon:
            print("terminado num de interaciones:" + str(i+1))
            print(actualmin)
            return actualmin

        if fdx(actualmin, dx) * fdx(x1, dx) < 0:
            x2 = actualmin
        else:
            x1 = actualmin
        print("iteracion no " + str(i + 1))
        print(actualmin)

    return actualmin


# punto inicial
x1 = 40.0
x2 = 90.0
e = math.exp(1) ** -3
%time xmin = Seccant(x1, x2, fdx, e, 100)
print("x*: " + str(xmin))
print("f(x*):"  + str(f(xmin)))
print("f(x'): " + str(fdx(xmin, 0.001)))
```

# <a id="ajuste"> Ajuste polinomial cúbico</a>

```python
import numpy as np
import math

f = lambda x: 204165.5/(330 - 2 * x) + 10400/(x - 20)
fdx = lambda x,dx: (f(x+dx) - f(x-dx))/(2*dx)

def cubic(f,fdx,a,b,iter,epsilon):
    dx = 0.01
    x_min = 0
    for i in range(1, iter):

        fda = fdx(a,dx)
        fdb = fdx(b,dx)

        z = 3 * (f(a) - f(b)) / (b - a) + fda + fdb
        w = ((b - a) / np.abs(b - a)) * np.sqrt(z * z - fda * fdb)
        miu = calculateMiu(fda,fdb,w,z)
        print('iter ' + str(i) + '  a:' + str(a) + ' b:' + str(b))

        if miu>0 and miu<1:
            x_min = b - miu * (b - a)
        elif miu<0:
            x_min = b
        else:
            x_min = a

        if(abs(fdx(x_min,dx))<epsilon):
            break

        else:
            if(fda*fdx(x_min,dx)<0):
                b = x_min
            else:
                a = x_min

    print("x* =", str(x_min) + "  f(x*)=" + str(f(x_min)))
    return x_min




def calculateMiu(fda,fdb,w,z):
    miu = (fdb + w - z) / (fdb - fda + 2 * w)
    return  miu





epsilon = math.exp(1) ** -3
%time xmin = cubic(f,fdx, 40, 90,100,epsilon)
print("x*: " + str(xmin))
print("f(x*):"  + str(f(xmin)))
print("f(x'): " + str(fdx(xmin, 0.001)))
```

# <a id="dorada">Sección Dorada</a>

```python
import  math

f = lambda x: 204165.5/(330-2*x) + 10400/(x-20)
tau = 0.381967

def GoldenSection(a,b,epsilon,tau,iter):

    for i in range(1,iter):
        alpha1 = a * (1 - tau) + b * tau
        alpha2 = a * tau + b * (1 - tau)
        fa1 = f(alpha1)
        fa2 = f(alpha2)

        if (fa1 > fa2):
            a = alpha1
        else:
            b = alpha2

        if abs(fa1 - fa2) < epsilon:
            break
        print('a={0}, b={1},  U(T*)={2} ite={3} \n'.format(a, b, f(a), i))
    print('T*={0}  U(T*)= {1} iteraciones={2} \n'.format(a, f(a), i))
    return a


e = math.exp(1) ** -10


%time xmin = GoldenSection(40,90,0.0001,tau,100)
print("x*: " + str(xmin))
print("f(x*):"  + str(f(xmin)))
print("f(x'): " + str(fdx(xmin, 0.001)))
```
