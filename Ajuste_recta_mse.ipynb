```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, 0].values #Feature matrix
y = dataset.iloc[:, 1].values #Criterion Matrix 
names = dataset.columns
```

```python
N=len(x)
sumxy=np.sum(x*y)
sumy=np.sum(y)
sumx=np.sum(x)
sumx2=sumx**2
sumxx=np.sum(x**2)
m= (N*sumxy - sumy*sumx)/(N*sumxx- sumx2)
b=(sumy*sumxx-sumxy*sumx)/(N*sumxx- sumx2)

#plot 
x_p=np.array([np.min(x),np.max(x)])
y_p=m*x_p+b
plt.plot(x_p,y_p,'r-')
plt.plot(x,y,'b.')
plt.xlabel(names[0], fontsize=12);
plt.ylabel(names[1], rotation=90, fontsize=12);
plt.title('model: y=%.2f x+ %.2f'% (m, b));
```
