```python
import math
import numpy as np
import matplotlib.pyplot as plt


#variable aleatoria
xt = np.arange(0,1,0.001)

#funcion analitica 
f  = lambda  x: 2*np.sin(5*x) + (3*(x **3))-(2*(x**2)) + 3*x - 5
#primera derivada
fd = lambda x: 10*np.cos(5*x) + (9*(x**2)) - (4*(x))+3
#segunda derivada
fdd = lambda x: (-1*(50 * np.sin(5*x)) + (18*x)-4)




#aproximada de primera derivada 
afda = lambda x, dx: ((f(x+(dx*x))-f(x))/dx*x)
#atrasada
afdt = lambda x, dx: ((f(x)-f(x-x*dx))/dx*x)
#centrada
afdc = lambda x,dx: (f(x+(dx*x))-f(x-(dx*x)))/dx*(x**2)

#aproximada de segunda derivada 
afdda = lambda x, dx: (f(x+(dx*x))-2*f(x) +f(x-(dx*x)))/dx**2

dx=0.01
fx = f(xt)
fp_adelantada=(afda(xt,dx)) 
fp_atrasada=(afdt(xt,dx))
fp_centrada = (afdc(xt,dx))

fpp_adelantada=(afdda(xt,dx))





g= plt.figure(2)
g.suptitle('f\'(x) Aproximada', fontsize=12)
plt.plot(xt,fx, label="f(x)", color="red")
plt.plot(xt,fp_adelantada, label="Adelantada")
plt.plot(xt,fp_atrasada, label = "Atrasada")
plt.plot(xt,fp_centrada,label = "Centrada")
plt.grid(True)
plt.legend(loc='upper left', frameon=False)
plt.show()

h= plt.figure(3)
h.suptitle('f\'\'(x) Aproximada', fontsize=12)
plt.plot(xt,fx, label="f(x)", color="red")
plt.plot(xt,fpp_adelantada, label="segunda derivada")
plt.grid(True)
plt.legend(loc='upper left', frameon=False)
plt.show()

```
