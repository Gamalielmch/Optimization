```python
from time import time
```

```python
U =lambda t:((204165.5)/(330-2*t))+(10400/(t-20))
dx = lambda x, delta: (U(x+delta)-U(x))/delta
```

```python
x1=40
x2=90
tau=0.618034
epsilon=0.0000001
start_time = time()

for i in range(30):
    x3=x1*tau +x2*(1-tau)
    x4=x1*(1-tau) +x2*tau
    f3=U(x3)
    f4=U(x4)
    if f3<f4:
        x2=x4
    else:
        x1=x3
    if abs(f4- f3)<epsilon:
        break
x=(x1+x2)/2
elapsed_time = time() - start_time
elapsed_time = elapsed_time*1000
print("Tiempo transcurrido: %f ms." % elapsed_time )
print("x: %f" % x)
print("fx: %f" % U(x))
print("f'x: %f" % dx(x,0.01))
print("iteraciones: %d" % i)
```
